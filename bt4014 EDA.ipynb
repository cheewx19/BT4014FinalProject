```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime as dt
df = pd.read_csv('NEER_Economic_Indicators - main.csv',parse_dates = ['date'], dayfirst=True)
```

```python
df.head()
```

```python
for var in list(df.columns[1:]):
    plt.title(var)
    plt.plot("date", var, data = df)
    plt.show()
```

```python
df['year'] = pd.DatetimeIndex(df['date']).year
```

```python
df['month'] = pd.DatetimeIndex(df['date']).month
```

```python
df.groupby('year').mean()['neer'].plot(kind='line')
```

```python
for var in list(df.columns[1:10]):
    plt.title(var)
    df.groupby('year').mean()[var].plot(kind='line')
    plt.show()
```

```python
for var in list(df.columns[1:10]):
    plt.title(var)
    df.groupby('month').mean()[var].plot(kind='line')
    plt.show()
```

```python
import seaborn as sns
for var in list(df.columns[1:10]):
    plt.title(var)
    sns.distplot(df[var], hist=True, kde=True,
             bins=20,              
             color = 'blue', 
             hist_kws={'edgecolor':'black'},
             kde_kws={'linewidth': 2})
    plt.show()
```
